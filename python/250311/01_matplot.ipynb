```python
# 데이터 로드 
drinks = pd.read_csv("../../csv/drinks.csv")
```

```python
drinks.head()
```

```python
plt.scatter(
    drinks['beer_servings'], 
    drinks['spirit_servings']
)
plt.show()
```

```python
# 데이터에서 상관계수를 출력 
#  corr()함수를 이용
drinks_corr = drinks.iloc[:, 1:5].corr(method='pearson')
drinks_corr
```

```python
# 히트맵 출력 
cols_view = ['beer', 'spirit', 'wine', 'alcohol']

hm = sns.heatmap(
    drinks_corr.values, 
    cbar = True,   # 사이드바 유무
    annot = True,  # 히트맵에 숫자 표시 유무
    square= False,  # 히트맵을 정사각형 생성 유무
    fmt = '.2f',   # 숫자의 소수점 자리수
    annot_kws= {'size' : 20}, # 숫자의 크기 
    xticklabels= cols_view, 
    yticklabels= cols_view, 
    cmap = 'Greens'
)

plt.show()
```

```python
# 데이터의 분포도 확인 
sns.pairplot(
    drinks.loc[ : , 'beer_servings' : 'total_litres_of_pure_alcohol'],
    height = 2.5, 
    corner=True
)
```
